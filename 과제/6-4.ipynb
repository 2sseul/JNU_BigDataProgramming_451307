## 컴퓨터공학과 2025208017 김이슬

```python
import seaborn as sns

anscombe = sns.load_dataset('anscombe')

# III번 데이터셋
dataset_III = anscombe[anscombe['dataset'] == 'III']
print("[Dataset III]")

print("데이터 수 (count):")
print(dataset_III.describe().loc['count'])

print("\n평균 (mean):")
print(dataset_III.describe().loc['mean'])

print("\n표준편차 (std):")
print(dataset_III.describe().loc['std'])

print("\n최솟값 (min):")
print(dataset_III.describe().loc['min'])

print("\n최댓값 (max):")
print(dataset_III.describe().loc['max'])

print("\n-----------------------------------\n")

# IV번 데이터셋
dataset_IV = anscombe[anscombe['dataset'] == 'IV']
print("[Dataset IV]")

print("데이터 수 (count):")
print(dataset_IV.describe().loc['count'])

print("\n평균 (mean):")
print(dataset_IV.describe().loc['mean'])

print("\n표준편차 (std):")
print(dataset_IV.describe().loc['std'])

print("\n최솟값 (min):")
print(dataset_IV.describe().loc['min'])

print("\n최댓값 (max):")
print(dataset_IV.describe().loc['max'])
```
