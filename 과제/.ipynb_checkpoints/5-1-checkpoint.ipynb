## 컴퓨터공학과 2025208017 김이슬

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 파라메트릭 하트 곡선 정의
t = np.linspace(0, 2 * np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

# 이동된 하트 곡선 좌표
x1, y1 = x - 20, y - 20  # Red Heart
x2, y2 = x + 20, y + 20  # Blue Heart

# 그래프 그리기
plt.figure(figsize=(8, 8))
plt.plot(x1, y1, color='red', label='Red Heart')
plt.plot(x2, y2, color='blue', label='Blue Heart')
plt.xlim(-40, 40)
plt.ylim(-40, 40)
plt.legend()
plt.title("Translated Parametric Hearts")
plt.show()
```
