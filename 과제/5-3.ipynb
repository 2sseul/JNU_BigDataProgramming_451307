```python
import numpy as np
import matplotlib.pyplot as plt

# subplot을 위한 표준편차 설정
std_list = [
    [1.0] * 10,
    [1.5] * 10,
    [2.0] * 10,
    [2.5] * 10  
]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for i, ax in enumerate(axes.flat):
    datasets = []
    for std in std_list[i]:
        data = np.random.normal(loc=0, scale=std, size=100)
        datasets.append(data)

    ax.boxplot(datasets)
    ax.set_ylim(-4, 4)
    ax.set_xticks(range(1, 11))
    ax.set_xticklabels([f'Data {j}' for j in range(1, 11)], rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
