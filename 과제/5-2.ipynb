## 컴퓨터공학과 2025208017 김이슬

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 난수 고정
np.random.seed(2025)

# 1부터 10까지의 정수 중 무작위로 100개 생성
data = np.random.randint(1, 11, size=100)

# 히스토그램 그리기
plt.figure(figsize=(8, 6))
plt.hist(data, bins=np.arange(1, 12))
plt.xticks(range(1, 11))
plt.show()
```
