## 6-3

```python
import numpy as np
```

```python
patients = np.array([[1.63, 1.86, 1.63, 1.76],
                     [76.0, 84.0, 69.0, 85.0]])

heights = patients[0]
weights = patients[1]

BMI = weights / (heights ** 2)

print("환자들의 키 :", heights)
print("환자들의 몸무게 :", weights)
print("환자들의 체질량지수 :", BMI)
print("체질량지수가 25미만 값 :", BMI[ BMI < 25])
```
