```python
height_mu = 194
height_sigma = 11
weight_mu = 78
weight_sigma = 8

player_num = 200
players = np.zeros((2, player_num))

players[0] = height_mu + height_sigma * np.random.randn(player_num)
players[1] = weight_mu + weight_sigma * np.random.randn(player_num)

print(np.mean(players[0]))
print(np.median(players[0]))
print(np.mean(players[1]))
print(np.median(players[1]))
```

```python
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
ranks = [5,3,2,3,4,6,7,4]

# 그래프 그리기 plt.plot(x축데이터, y축데이터, 선의 색상, 마커 정보)
plt.plot(years, ranks, color='blue', marker='*')

plt.title("Tottenham Hotspur FC Ranking")

plt.xlabel("Year")
plt.ylabel("Ranking")

plt.show()
```

```python
x = list(range(-20, 20))

y1 = [2*t for t in x]
y2 = [t**2 for t in x]
y3 = [-.2*t**2 -5 for t in x]

plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.axis([-30, 30, -30, 30])
```

```python
x = list(range(-20, 20))

y1 = [2*t for t in x]
y2 = [t**2 for t in x]
y3 = [-.2*t**2 -5 for t in x]

plt.plot(x, y1, 'r--', label='2*x')
plt.plot(x, y2, 'g^-', label='x^2', marker='^')
plt.plot(x, y3, 'b*:', label='-0.2^2-5', marker='*')
plt.axis([-30, 30, -30, 30])
plt.legend()
```

```python
x = np.arange(-20, 20)

y1 = x
y2 = 3*x
y3 = x**2-5

plt.plot(x, y1, 'r--', label='x')
plt.plot(x, y2, 'g-', label='3^x')
plt.plot(x, y3, 'b:', label='x^2-5')
plt.axis([-20, 20, -20, 100])
plt.legend(loc='lower right')
```
