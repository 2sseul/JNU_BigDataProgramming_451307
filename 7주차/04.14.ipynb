## 5-1

```python
import matplotlib.pyplot as plt

x = list(range(-10, 10)) ## -10에서 10 사이의 수를 1간격으로 생성
y = [2*t for t in x]  ## 2*x를 원소로 가지는 y 함수
plt.plot(x, y, marker='o')

plt.axis([-20, 20, -20, 20]) ## 그림을 그릴 영역 지정
plt.show()
```

## 5-4

```python
import numpy as np

x = np.arange(0, 60)
y1 = x
y2 = 2 * x
y3 = x ** 2

plt.plot(x, y1, 'r--', label='y=x')
plt.plot(x, y2, 'g-', label='y=2*x')
plt.plot(x, y3, 'b:', label='y=x**2')

plt.axis([-2, 60, -20, 90])
plt.legend()
```
