```python
import seaborn as sns

# 테마를 저장하는 기능이 있음
sns.set_theme(style="darkgrid")
# 시본에서 제공하는 팁 데이터를 가져오자
tips = sns.load_dataset("tips")
# 팁 데이터의 시각화 기능을 호출하자
sns.relplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker", style="smoker", size="size")
```

```python
sns.histplot(tips['tip'], kde=False, bins=10) # histplot() => 데이터의 분포를 히스토그램으로 나타내는 함수
```

```python
# bins를 12로 늘리면 가로축이 촘촘하게(세밀하게) 분할된 것을 볼 수 있음
sns.histplot(tips['tip'], kde=False, bins=12)
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
sns.histplot(tips['tip_pct'], kde=True, bins=20) 
# kde = True시 가우시안 커널 밀도 추정을 실선으로 나타내주는 역할
# kde로 이상치 데이터를 탐색할 수 있다.
# 밀도 기반 이상치 탐지법이란 데이터의 분포를 사용하여 이상치를 나타내는 방법
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
# 식사 요금과 팁과의 상관관계 y: 팁 비용 / x: 전체 식사 비용
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='smoker', data=tips)
```

```python
sns.pairplot(tips)
```
